```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
pd.get_dummies(data)
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data['whoAmI_human'] = pd.Series(['True' if i == 'robot' else 'False' for i in data['whoAmI']])
data['whoAmI_robot'] = pd.Series(['False' if i == 'robot' else 'True' for i in data['whoAmI']])
print(data)
```
